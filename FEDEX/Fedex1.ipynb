```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

data = {
    "case_id": range(1, 101),
    "invoice_amount": np.random.randint(5000, 500000, 100),
    "overdue_days": np.random.randint(1, 180, 100),
    "customer_risk": np.random.choice(["Low", "Medium", "High"], 100),
    "past_recovery_rate": np.random.uniform(0.2, 0.95, 100)
}

df = pd.DataFrame(data)
df.head()
```

```python
risk_map = {"Low": 0.3, "Medium": 0.6, "High": 1.0}
df["customer_risk_score"] = df["customer_risk"].map(risk_map)
```

```python

def normalize(col):
    return (col - col.min()) / (col.max() - col.min())

df["norm_overdue"] = normalize(df["overdue_days"])
df["norm_invoice"] = normalize(df["invoice_amount"])

df["priority_score"] = (
    0.35 * df["norm_overdue"] +
    0.30 * df["norm_invoice"] +
    0.20 * df["customer_risk_score"] +
    0.15 * (1 - df["past_recovery_rate"])
)

df["priority_bucket"] = pd.qcut(
    df["priority_score"],
    q=[0, 0.3, 0.7, 1.0],
    labels=["Low", "Medium", "High"]
)

df["priority_bucket"].value_counts()
```

```python
def urgency_band(days):
    if days >= 120:
        return "Critical"
    elif days >= 60:
        return "High"
    elif days >= 30:
        return "Medium"
    else:
        return "Low"

df["urgency_band"] = df["overdue_days"].apply(urgency_band)
df["urgency_band"].value_counts()
```

```python
df[["case_id", "priority_bucket", "urgency_band"]].head(10)
```

```python
df["priority_bucket"].value_counts().plot(kind="bar")
plt.title("AI-Based Case Priority Distribution")
plt.xlabel("Priority Level")
plt.ylabel("Number of Cases")
plt.show()

df["urgency_band"].value_counts().plot(kind="bar")
plt.title("Overdue Case Urgency Levels")
plt.xlabel("Urgency Band")
plt.ylabel("Number of Cases")
plt.show()
```

```python
df_sorted = df.sort_values(
    ["priority_score", "overdue_days"],
    ascending=False
)

df_sorted.head(10)
```

```python
# Urgency bands based on overdue days
df["urgency_band"] = pd.cut(
    df["overdue_days"],
    bins=[0, 30, 60, 90, 180],
    labels=["Low", "Medium", "High", "Critical"]
)

# Operational action based on priority
df["allocation_action"] = np.where(
    df["priority_bucket"] == "High",
    "Immediate DCA Allocation",
    np.where(
        df["priority_bucket"] == "Medium",
        "Standard Follow-up",
        "Monitor"
    )
)

df[[
    "case_id",
    "overdue_days",
    "urgency_band",
    "priority_bucket",
    "allocation_action"
]].head(10)
```

```python
# Recovery Probability Logic
df["recovery_probability"] = (
    0.4 * df["past_recovery_rate"] +
    0.3 * (1 - df["norm_overdue"]) +
    0.3 * (1 - df["customer_risk_score"])
)

df["recovery_probability"] = df["recovery_probability"].clip(0, 1)

df[["case_id", "recovery_probability"]].head()

# Recovery Buckets
df["recovery_bucket"] = pd.cut(
    df["recovery_probability"],
    bins=[0, 0.4, 0.7, 1.0],
    labels=["Low", "Medium", "High"]
)

print(df["recovery_bucket"].value_counts())

# Recovery vs Overdue
plt.scatter(df["overdue_days"], df["recovery_probability"], alpha=0.6, color="maroon")
plt.xlabel("Overdue Days")
plt.ylabel("Recovery Probability")
plt.title("Recovery Probability vs Overdue Days")
plt.show()
```

```python
# SLA Risk Score
df["sla_risk_score"] = (
    0.5 * df["norm_overdue"] +
    0.3 * df["priority_score"] +
    0.2 * (1 - df["past_recovery_rate"])
)

# SLA Risk Buckets
df["sla_risk_bucket"] = pd.cut(
    df["sla_risk_score"],
    bins=[0, 0.4, 0.7, 1.0],
    labels=["Low", "Medium", "High"]
)
print(df["sla_risk_bucket"].value_counts())
```

```python
# SLA Risk Distribution
df["sla_risk_bucket"].value_counts().plot(kind="bar")
plt.title("SLA Breach Risk Distribution")
plt.xlabel("SLA Risk Level")
plt.ylabel("Number of Cases")
plt.show()
```

```python
df = df[[
    "case_id",
    "recovery_probability",
    "recovery_bucket",
    "sla_risk_bucket"
]]

print(df.head(10))

''' DCA Performance Analytics & System Insights
This section evaluates Debt Collection Agent (DCA) performance using AI-generated
priority, recovery probability, and SLA risk indicators.'''

# Simulate DCA assignment
np.random.seed(10)
df["dca_id"] = np.random.choice(
    ["DCA_1", "DCA_2", "DCA_3", "DCA_4"],
    size=len(df)
)

print(df[["case_id", "dca_id"]].head())
```

```python


# Simulated recovery outcome
df["recovered"] = np.random.binomial(
    1, df["recovery_probability"]
)

# SLA breach assumed for high SLA risk
df["sla_breached"] = np.where(
    df["sla_risk_bucket"] == "High", 1, 0
)

df[["case_id", "recovered", "sla_breached"]].head()


dca_metrics = df.groupby("dca_id").agg(
    total_cases=("case_id", "count"),
    recovered_cases=("recovered", "sum"),
    avg_recovery_probability=("recovery_probability", "mean"),
    sla_breaches=("sla_breached", "sum")
).reset_index()

dca_metrics["recovery_rate"] = (
    dca_metrics["recovered_cases"] / dca_metrics["total_cases"]
)

dca_metrics["sla_compliance_rate"] = (
    1 - dca_metrics["sla_breaches"] / dca_metrics["total_cases"]
)

dca_metrics


dca_metrics["performance_score"] = (
    0.4 * dca_metrics["recovery_rate"] +
    0.3 * dca_metrics["sla_compliance_rate"] +
    0.3 * dca_metrics["avg_recovery_probability"]
)

dca_metrics.sort_values("performance_score", ascending=False)


plt.figure()
plt.bar(
    dca_metrics["dca_id"],
    dca_metrics["performance_score"]
)
plt.title("DCA Performance Leaderboard")
plt.xlabel("DCA")
plt.ylabel("Performance Score")
plt.show()
```

```python

'''Key Insights
- High-priority cases with low recovery probability require escalation.
- DCA performance varies significantly, justifying AI-based workload allocation.
- SLA risk indicators enable proactive intervention before breaches occur.'''


'''Conclusion
This system demonstrates how AI-driven prioritization, recovery prediction,
and performance analytics can improve debt collection efficiency.'''
```
